# A. Librerias y rutas

```python
import re

# Ruta del archivo
file_path = "./results/codigos.json"

# Leer el archivo completo en memoria línea por línea
with open(file_path, "r", encoding="utf-8") as file:
    lineas = file.readlines()  # Guarda todas las líneas en una lista

# Lista para almacenar los números de línea donde aparece "codigo"
lineas_codigo = []

# Buscar líneas que contienen "codigo"
for num_linea, line in enumerate(lineas, start=0):
    if "codigo" in line:
        lineas_codigo.append(num_linea)

# Imprimir las líneas donde aparece "codigo"
print("Líneas con 'codigo':", lineas_codigo)
```

```python
# Buscar líneas que contienen "codigo"
codigos = []
for num_linea, line in enumerate(lineas, start=0):
    if "codigo" in line:
        # Encontrar las posiciones de "codigo:" y "\n"
        pos_inicio = line.find("codigo:") + len("codigo:")
        pos_fin = line.find("\n")
        
        # Extraer el substring y limpiarlo de comillas y espacios
        codigo = line[pos_inicio:pos_fin].strip().strip('"')
        codigos.append(codigo)

# Imprimir los códigos extraídos
print("Códigos encontrados:", codigos)
```

```python
# guardar los codigos en un txt
with open("./results/codigos.txt", "w") as file:
    for codigo in codigos:
        file.write(codigo + "\n")
```

```python
# a cada codigo le agrego el link de la pagina de senamhi https://www.senamhi.gob.pe/site/descarga-datos/descarga/?cod={}
# Crear las URLs completas
urls = []
base_url = "https://www.senamhi.gob.pe/site/descarga-datos/descarga/?cod={}"
for codigo in codigos:
    url = base_url.format(codigo)
    urls.append(url)

# Guardar las URLs en un archivo
with open("./results/urls.txt", "w") as file:
    for url in urls:
        file.write(url + "\n")

print(f"Se generaron {len(urls)} URLs y se guardaron en ./results/urls.txt")
```
